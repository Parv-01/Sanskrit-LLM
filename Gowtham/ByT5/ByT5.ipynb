```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
from tqdm import tqdm

# -----------------------------
# Load model
# -----------------------------
model_name = "buddhist-nlp/byt5-sanskrit"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def translate_batch(text_list):
    inputs = [
        f"Translate the following English medical description into Sanskrit in IAST format: {str(t)}"
        for t in text_list
    ]

    tokenized = tokenizer(
        inputs,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=128
    ).to(device)

    with torch.no_grad():
        outputs = model.generate(
            **tokenized,
            max_length=80,
            num_beams=4,
            do_sample=False,
            early_stopping=True
        )

    decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    decoded = [d.strip() for d in decoded]

    return decoded

    with torch.no_grad():
        outputs = model.generate(
            **tokenized,
            max_length=80,
            num_beams=4,
            do_sample=False,
            early_stopping=True
        )

    decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    decoded = [d.strip() for d in decoded]

    return decoded


# -----------------------------
# Apply on DataFrame (batching)
# -----------------------------
batch_size = 16   # adjust based on GPU (8 if OOM)

results = []

texts = df_gsheet["English_meaning"].fillna("").tolist()

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i+batch_size]
    translated = translate_batch(batch)
    results.extend(translated)

# -----------------------------
# Store output column
# -----------------------------
df_gsheet["ByT5_Output"] = results

# Preview
df_gsheet.head()
```
