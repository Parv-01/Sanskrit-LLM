```python
import gspread
from google.colab import auth
import pandas as pd
from google.auth import default

# Authenticate Google Colab to access your Google Drive and Sheets
auth.authenticate_user()

# Get the credentials authenticated by Colab
creds, project = default()

# Initialize gspread client with these credentials
gc = gspread.Client(auth=creds)

# Replace 'DATASET' with the exact name of your Google Sheet
sheet_name = 'DATASET' # The main spreadsheet name
worksheet_name = 'KG_Data' # The specific worksheet name

try:
    # Open the sheet by its name
    spreadsheet = gc.open(sheet_name)
    # Select the specific worksheet by its name
    worksheet = spreadsheet.worksheet(worksheet_name)

    # Get all values from the worksheet as a list of lists
    data = worksheet.get_all_values()

    # Convert the list of lists to a pandas DataFrame
    df_kg_data = pd.DataFrame(data[1:], columns=data[0]) # First row as headers

    print(f"Successfully loaded the Google Sheet '{sheet_name}' and worksheet '{worksheet_name}' into a DataFrame.")
    display(df_kg_data.head())

except gspread.exceptions.SpreadsheetNotFound:
    print(f"Error: Google Sheet named '{sheet_name}' not found. Please ensure the name is exact (case-sensitive) and you have access permissions.")
except gspread.exceptions.WorksheetNotFound:
    print(f"Error: Worksheet named '{worksheet_name}' not found in spreadsheet '{sheet_name}'. Please ensure the worksheet name is exact (case-sensitive).")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import json
import pandas as pd

verse_data = []

# Iterate through each row of the DataFrame
for index, row in df_kg_data.iterrows():
    # Process 'Symptoms of disease (Verse numbers)'
    symptoms_verse = row['Symptoms of disease (Verse numbers)']
    if pd.notna(symptoms_verse) and symptoms_verse.strip() != '':
        verse_data.append({
            'Verse_Number': symptoms_verse,
            'Verse_Type': 'Symptoms'
        })

    # Process 'Nature of disease (Verse numbers)'
    nature_verse = row['Nature of disease (Verse numbers)']
    if pd.notna(nature_verse) and nature_verse.strip() != '':
        verse_data.append({
            'Verse_Number': nature_verse,
            'Verse_Type': 'Nature'
        })

    # Process 'Treatment of disease (Verse numbers)'
    treatment_verse = row['Treatment of disease (Verse numbers)']
    if pd.notna(treatment_verse) and treatment_verse.strip() != '':
        verse_data.append({
            'Verse_Number': treatment_verse,
            'Verse_Type': 'Treatment'
        })

# Remove duplicates from verse_data
unique_verse_data = []
seen = set()
for entry in verse_data:
    # Convert dictionary to a tuple of (key, value) pairs for hashing
    entry_tuple = tuple(sorted(entry.items()))
    if entry_tuple not in seen:
        seen.add(entry_tuple)
        unique_verse_data.append(entry)

# Define the output path for the JSON file in Google Drive
output_verse_json_path = '/content/drive/MyDrive/Sanskrit_LLM/Dataset/Sample_KG_Data/Verse_Node_Info.json'

# Save the unique data to a JSON file
with open(output_verse_json_path, 'w', encoding='utf-8') as f:
    json.dump(unique_verse_data, f, ensure_ascii=False, indent=4)

print(f"Unique verse information successfully saved to {output_verse_json_path} as a JSON file.")

# Display the first few entries of the JSON data to confirm
print("\nPreview of the saved JSON data:")
print(json.dumps(unique_verse_data[:5], ensure_ascii=False, indent=4))
```

```python
import json
import pandas as pd

# Initialize an empty set to store unique symptoms
unique_symptoms = set()

# Iterate through the 'SYMPTOMS' column
for symptoms_str in df_kg_data['SYMPTOMS']:
    if pd.notna(symptoms_str) and symptoms_str.strip() != '':
        # Split the string by comma and strip whitespace from each symptom
        symptoms_list = [symptom.strip() for symptom in symptoms_str.split(',')]
        # Add each symptom to the set
        for symptom in symptoms_list:
            if symptom:
                unique_symptoms.add(symptom)

# Convert the set of unique symptoms to a list of dictionaries
symptom_json_data = [{'Symptom_Name': symptom} for symptom in sorted(list(unique_symptoms))]

# Define the output path for the JSON file in Google Drive
output_symptom_json_path = '/content/drive/MyDrive/Sanskrit_LLM/Dataset/Sample_KG_Data/Symptom_Node_Info.json'

# Save the data to a JSON file
with open(output_symptom_json_path, 'w', encoding='utf-8') as f:
    json.dump(symptom_json_data, f, ensure_ascii=False, indent=4)

print(f"Unique symptom names successfully saved to {output_symptom_json_path} as a JSON file.")

# Display the first few entries of the JSON data to confirm
print("\nPreview of the saved JSON data:")
print(json.dumps(symptom_json_data[:5], ensure_ascii=False, indent=4))
```

```python
import json

# Your JSON file path
input_path = '/content/drive/MyDrive/Sanskrit_LLM/Dataset/Sample_KG_Data/Disease_Node_Info.json'

with open(input_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

queries = []

queries = []

for item in data:

    definition = item.get("Long_definition", "").replace('"', '\\"')

    query = f"""
MERGE (d:Disease {{
  name: "{item.get('NAMC_term', '')}",
  dev_name: "{item.get('NAMC_term_DEVANAGARI', '')}",  #DEVANAGARI SCRIPT
  diacritical: "{item.get('NAMC_term_diacritical', '').replace('(a) ', '')}",  #diacritical script
  english_name: "{item.get('Name English', '')}",
  namc_id: "{item.get('NAMC_ID', '')}",
  namc_code: "{item.get('NAMC_CODE', '')}",
  definition: "{definition}",
  project: "Sanskrit_LLM"
}});
"""
    queries.append(query)
```

```python
queries = []

for _, row in df_kg_data.iterrows():
    disease = row.get("NAMC_term_DEVANAGARI", "").strip()
    symptoms_field = row.get("SYMPTOMS", "").strip()   # <-- column name

    if disease and symptoms_field:

        # Split by comma
        symptoms = symptoms_field.split(",")

        for sym in symptoms:
            sym = sym.strip().replace('"', '\\"')

            if sym:
                query = f"""
MATCH (d:Disease {{dev_name:"{disease}"}})
MATCH (s:Symptom {{name:"{sym}"}})
MERGE (d)-[:HAS_SYMPTOM]->(s);
"""
                queries.append(query)

print("✅ Total symptom edges:", len(queries))
```

```python
import json

# Input file path
input_path = '/content/drive/MyDrive/Sanskrit_LLM/Dataset/Sample_KG_Data/Verse_Node_Info.json'

# Load JSON
with open(input_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

queries = []

for item in data:

    verse_number = item.get("Verse_Number", "").replace('"', '\\"').strip()
    verse_type = item.get("Verse_Type", "").replace('"', '\\"').strip()

    if verse_number and verse_type:
      query = f"""
MERGE (v:Verse {{
  verse_range: "{verse_number}",
  type: "{verse_type}"
}})
SET v.project = "Sanskrit_LLM",
    v.display = "{verse_number} ({verse_type})";
"""
      queries.append(query)

# Save queries to file
output_path = '/content/drive/MyDrive/Sanskrit_LLM/Dataset/Sample_KG_Data/Verse_Nodes_Queries.txt'

with open(output_path, 'w', encoding='utf-8') as f:
    f.write("\n".join(queries))

print("✅ Verse node queries saved at:", output_path)
```
