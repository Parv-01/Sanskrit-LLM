```python
def translate_batch_qwen(text_list):

    iast_outputs = []
    devanagari_outputs = []

    for text in text_list:

        prompt = (
            "You are an Ayurvedic Medical Scholar. Provide ONLY the technical IAST term. "
            "No notes, no explanations.\n\n"
            "English: facial paralysis -> IAST: ardita\n"
            "English: excessive thirst -> IAST: tṛṣṇā\n"
            f"English: {text} -> IAST:"
        )

        inputs = tokenizer([prompt], return_tensors="pt").to(model.device)

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=10,
                temperature=0.0,
                do_sample=False
            )

        response = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]

        # Extract IAST
        raw_output = response.split("IAST:")[-1].strip().split("\n")[0]

        # Clean IAST (VERY IMPORTANT)
        clean_iast = re.sub(
            r'[^a-zA-Zāīūṛḷṅñṭḍṇśṣḥṃ\-\s]',
            '',
            raw_output
        ).split(' ')[0].strip()

        # Convert to Devanagari
        try:
            devanagari = transliterate.process("IAST", "Devanagari", clean_iast)
        except:
            devanagari = ""

        iast_outputs.append(clean_iast)
        devanagari_outputs.append(devanagari)

    return iast_outputs, devanagari_outputs


```

```python
batch_size = 4   # keep small for stability

texts = df_gsheet["English_meaning"].fillna("").astype(str).tolist()

iast_results = []
dev_results = []

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i+batch_size]

    iast_batch, dev_batch = translate_batch_qwen(batch)

    iast_results.extend(iast_batch)
    dev_results.extend(dev_batch)


# -----------------------------
# STORE OUTPUT
# -----------------------------
df_gsheet["Qwen_IAST_OP"] = iast_results
df_gsheet["Qwen_Devanagari_OP"] = dev_results

df_gsheet.head()
```
